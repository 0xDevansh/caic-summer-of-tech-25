## Dataset Generator
In this notebook I'll generate a synthetic dataset to test the linear regressor using numpy.
The epsilon (random error) will be a random from a Gaussian disribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
def get_dataset(n_parameters, n_samples, coefs_scale=10, x_scale=10, y_scale=10, noise_scale=2):
    # First generate a linear function to determine our coefficients
    coefs = np.random.randn(n_parameters) * coefs_scale
    const = np.random.randn(1) * coefs_scale
    # Generate samples
    X = np.random.randn(n_samples, n_parameters)
    Y_pure = X @ coefs + const
    # Add noise
    Y = Y_pure + np.random.randn(n_samples) * noise_scale
    return X, Y
```

```python
# visualising a single line dataset
np.random.seed(42)
n_parameters = 1
n_samples = 100
X, Y = get_dataset(n_parameters, n_samples)
plt.scatter(X, Y)
```

```python
X, Y = get_dataset(4, 1000)
np.savetxt('X.npy', X)
np.savetxt('Y.npy', Y)
```
